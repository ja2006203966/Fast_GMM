```python
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

```python
def norm_dist(x, mu = 0, std=1):
    return np.exp(-((x-mu)/std)**2/2)/(np.sqrt(2*np.pi)*std)
def submerge(x, ans, E, i):
    E2 = set(E[i])&set(ans.keys())
    E2 = [j for j in E2 if (i in E[j])&(j!=i)]
    ans[i]['INDEX'] = ans[i]['INDEX']|set(E2)
    n1 = len(ans[i]['INDEX'])
    # ans[i]['MEAN'] = (ans[i]['MEAN']*n1+ np.sum([ans[e]['MEAN'] for e in E2 if e!=i]))/(n1+n2)
    # ans[i]['MEAN'] = (ans[i]['MEAN']*n1+ np.sum([np.mean(ans[e]['VALUES']) for e in E2 if e!=i]))/(n1+np.sum([len(ans[e]['VALUES']) for e in E2 if e!=i]))
    ans[i]['MEAN'] = np.mean(x[list(ans[i]['INDEX'])])
    ans[i]['VALUES'] = list(x[list(ans[i]['INDEX'])])
    for k in E2:
        if k!=i:
            # ans[i]['VALUES']+=ans[k]['VALUES']
            del ans[k]
    return ans
def merging(x, E, D):
    ans = {i:{'INDEX':set([i]), 'MEAN':v, 'VALUES':[v]} for i, v in enumerate(x)}
    keys = list(ans.keys())
    Etot = []
    for k in tqdm(keys):
        Etot+=list(E[k])
    idx, count = np.unique(Etot, return_counts=True)
    k2 = idx[count>1]
    # k2 = []
    # for i in tqdm(keys):
    #     for j in keys:
    #         if j==i:
    #             continue
    #         if i in E[j]:
    #             # n+=1
    #             k2.append(i)
    #             break
    # k2 = np.array(k2)
    _, n2 = np.unique(E, return_counts=True)
    n2 = (n2[k2]-1)
    idx = np.argsort(n2)[::-1] #assembling
    k2 = k2[idx]
    keys = k2
    for i in tqdm(keys):
        if i in ans.keys():
            ans = submerge(x, ans, E, i)
    return ans
def self_distance(x, n=2):
    # x: (n,)
    xd = np.expand_dims(x, axis = -1)
    R = xd*(np.transpose(xd, [1, 0])*0+1)
    RT = np.transpose(R, [1,0])
    D = np.abs(R-RT)
    mean, std = np.mean(D), np.std(D)
    Idx = np.argsort(D, axis = -1)
    E = Idx[:,:n+1]
    ans = merging(x, E, D)
    # E = np.indices(D.shape)[0][D<mean]
    # _, E = np.unique(E, return_counts=True)
    # E = np.expand_dims(E, axis=-1)
    # E = E*(np.transpose(E, [1, 0])*0+1)
    # W = D/E
    # Idx = np.argsort(W, axis = -1)
    # E = Idx[:,:10]
    # unique, counts = np.unique(E, return_counts=True)
    # ans = dict(zip(unique, counts))
    return ans
# def GMM(a, x):
#     y=0
#     for i in a:
#         if i[2]==0:
#             continue
#         y = y+i[0]*norm_dist(x, mu=i[1], std=i[2])
#     return y
class GMM:
    def __init__(self, a):
        self.a = a
    def GMM(self, x):
        y=0
        for i in a:
            if i[2]==0:
                continue
            y = y+i[0]*norm_dist(x, mu=i[1], std=i[2])
        return y
class Normal:
    def __init__(self, mu, std):
        self.mu = mu
        self.std = std
    def dist(self, x):
        return np.exp(-((x-self.mu)/self.std)**2/2)/(np.sqrt(2*np.pi)*self.std)
def integration(fn, xl, xu, N=10000):
    delta = (xu-xl)/N
    x = np.arange(xl, xu, delta)
    y = fn(x)
    return np.sum(delta*y)
def spec(a, n=5):
    lcl = []
    ucl = []
    for i in a:
        lcl.append(i[1]-n*i[2])
        ucl.append(i[1]+n*i[2])
    return min(lcl), max(ucl)
```

```python
n = 2000
x = np.random.normal(0, 1, n)
x = np.append(np.random.normal(5, 1, n), x, axis = 0)
x = np.append(np.random.normal(7, 3, n), x, axis = 0)
x = np.append(np.random.normal(10, 1, n), x, axis = 0)
```

```python
bins = 120
N = len(x)
edge_count = self_distance(x, n=int(N/10)+2)
fig, ax = plt.subplots(1,1, figsize = (8,8))
ax.hist(x, bins = bins, histtype = 'step', label = 'Raw')
xlim = ax.get_xlim()
xtot = np.array([])
a = []
for i in edge_count:
    x2 = np.array(sorted(edge_count[i]['VALUES']))
    xtot = np.append(xtot, x2)
    mu = edge_count[i]['MEAN']
    std = np.std(x2)
    n = len(edge_count[i]['VALUES'])/N
    a.append([n, mu, std])
    y = norm_dist(x2, mu = mu, std=std)*len(edge_count[i]['VALUES'])*(xlim[1]-xlim[0])/bins
    ax.plot(x2, y, label = "mu:{}, std:{}, n:{}%".format(round(mu,2), round(std,2), round(n*100,2)) )
    # ax.hist(x2, histtype = 'step', label = "mu:{}, std:{}, n:{}%".format(round(mu,2), round(std,2), round(len(edge_count[i]['VALUES'])/N*100,2)), bins = bins)
idx = np.argsort(xtot)
xtot = xtot[idx]
ytot = GMM(a).GMM(xtot)*N*(xlim[1]-xlim[0])/bins
ax.plot(xtot, ytot, '-', c='black', alpha=0.7)
lcl, ucl = spec(a, n=5)
ylim = ax.get_ylim()
ax.plot([lcl, lcl], [ylim[0], ylim[1]], '--', c='red', label = 'LCL')
ax.plot([ucl, ucl], [ylim[0], ylim[1]], '--', c='red', label = 'UCL')
ax.legend()
plt.show()
```
